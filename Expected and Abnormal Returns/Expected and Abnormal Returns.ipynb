### 4. Calculating expected returns (15 points)

Below is a dataframe (`df_erp`) that contains the daily equity risk premium (i.e. the market return minus the risk free return) of ten days in december last year. Suppose you are doing an event study and have calculated alphas and betas for two stocks (stock a and b), which you have stored in a list. Now you want to determine expected returns for these two stocks.

Write a function that takes as input `df_erp`, `beta_list`, `alpha_list` and `risk_free_rate` and calculates the expected return on each day for each stock using the capm. The output should be a dataframe with the columns as 'market', 'stock a', 'stock b', the dates as index, and the expected returns as rows. Call the dataframe `df_expected`.

Hint 1: Convert the list of alphas and betas into numpy arrays

Hint 2: The numpy function np.outer will likely come in handy

Hint 3: Remember expected return of a stock using the CAPM is calculated by multiplying the stock's beta by the equity risk premium and then adding the stocks alpha and risk free rate, i.e er = b* erp + alpha + rf.

```python
dates_2 = ['2-12-2019', '3-12-2019', '4-12-2019',
                 '5-12-2019', '6-12-2019', '9-12-2019',
                 '10-12-2019', '11-12-2019', '12-12-2019']

df_erp = pd.DataFrame([0.02, 0.04, 0.08, 0.02, 0.03, 0.10, 0.04, 0.06, 0.04],
         columns=['equity risk premium'],
          index= dates_2)

beta_list = [1, 1.1, 1.5]
alpha_list = [0, 0.015, 0.012]
risk_free_rate = 0.0072

#df_erp
df_erp
```

```python
def er_calculator(dataframe, beta_list, alpha_list, risk_free_rate):
# your code here
    alpha_array = np.array(alpha_list)
    beta_array = np.array(beta_list)    
    c = np.outer(dataframe['equity risk premium'], beta_array)
    
    
    df_expected = pd.DataFrame(data=np.add(c+risk_free_rate, alpha_array),
                               columns = ['market','stock a','stock b'],
                               index = dataframe.index)
    return df_expected

df_expected = er_calculator(df_erp, beta_list, alpha_list, risk_free_rate)
df_expected
```

### 5. Calculating abnormal returns (5 points)

Below is a dataframe (`df_actual`) that contains the actual (daily) returns of the market, stock a and stock b. Write a function that takes the previously created dataframe of expected (`df_expected`) returns and outputs a dataframe of abnormal returns for stocks a and b (thus dropping the column 'market'). Call the dataframe `df_abnormal`.


```python
dict_3 = {'market': [0.0272, 0.0472, 0.0872, 0.0272, 0.0372, 0.1072, 0.0472, 0.0672, 0.0472],
        'stock a': [0.0242, -0.0082, -0.489, 0.0185, 0.0766, 0.002, -0.0028, -0.003, -0.002],
          'stock b': [0.228, -0.0297, 0.0387, 0.0435, 0.079, 0.031, -0.058, 0.029, 0.002]
         }

df_actual = pd.DataFrame(dict_3, index= dates_2)

df_actual
```

```python
def ar_calculator(df_expected, df_actual):
    return pd.DataFrame(data=np.add(df_actual, -df_expected))[['stock a','stock b']]

df_abnormal = ar_calculator(df_expected, df_actual)
df_abnormal
```

### 6. Average abnormal returns (5 points)

Write a function that takes the previously created dataframe `df_abnormal` and creates a column called 'avg_ar', which contains the average abnormal returns of stock a and stock b for each date.


```python
def aar_calculator(df_abnormal):
    return pd.DataFrame(data=(df_abnormal['stock a']+df_abnormal['stock b'])/2,
                        columns = ['avg_ar'],
                        index = df_abnormal.index)

df_aar = aar_calculator(df_abnormal)
df_aar
```
